## Image Addition

We can either use traditional `numpy` addition operation or, `cv.add()` function to add two images. There is a difference between OpenCV addition and Numpy addition. OpenCV addition is a saturated operation while Numpy addition is a modulo operation.

For example, consider the below sample:

```python
x = np.uint8([[250]])
y = np.uint8([[10]])
```

```python
# opencv addition
cv.add(x, y)  # saturated at max_val of 255
```

```python
# numpy addition
x + y   # modulo operation, 250+10 = 260 % 256 = 4
```

 For the addition operation to work, both images should be of same depth and type, or the second image can just be a scalar value.

## Image Blending

This is also image addition, but different weights (**between 0 and 1**) are given to images in order to give a feeling of blending or transparency.

To do this we can use the opencv function, `cv.addWeighted(img1, w1, img2, w2, c)`.

```python
img1 = cv.resize(cv.imread('./imgs/castle.jpg'), (640, 640))
img2 = cv.resize(cv.imread('./imgs/star.jpeg'), (640, 640))

img3 = cv.addWeighted(img1, 0.6, img2, 0.2, 2)

plt.imshow(img3, 'brg_r')
```
