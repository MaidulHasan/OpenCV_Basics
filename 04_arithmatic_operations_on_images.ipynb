## **Image Addition**

We can either use traditional `numpy` addition operation or, `cv.add()` function to add two images. There is a difference between OpenCV addition and Numpy addition. OpenCV addition is a saturated operation while Numpy addition is a modulo operation.

For example, consider the below sample:

```python
x = np.uint8([[250]])
y = np.uint8([[10]])

# opencv addition
print("opencv addition: ", cv.add(x, y))  # saturated at max_val of 255

# numpy addition
print("numpy addition:", x + y)  # modulo operation, 250+10 = 260 % 256 = 4
```

This effect is more evident in actual image operations.

**For the addition operation to work, both images should be of same depth and type, or the second image can just be a scalar value.**

```python
img1 = cv.resize(
    cv.cvtColor(cv.imread("./imgs/castle.jpg"), cv.COLOR_BGR2RGB), (640, 640)
)
img2 = cv.resize(cv.imread("./imgs/star.jpeg"), (640, 640))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
plt.imshow(cv.add(img1, img2))
```

## **Image Blending**

This is also image addition, but different weights (**between 0 and 1**) are given to images in order to give a feeling of blending or transparency.

To do this we can use the opencv function, `cv.addWeighted(img1, w1, img2, w2, c)`.

```python
img3 = cv.addWeighted(img1, 0.6, img2, 0.2, 2)

plt.imshow(img3, "brg_r")
```

## **Bitwise Operations**

<img src='./imgs/bitwise_operations.png'>

Bitwise operations are really useful for putting logo in images and specially used for creating masks.

- ### Bitwise AND

```python
plt.imshow(cv.bitwise_and(img1, img2))
```

- ### Bitwise OR

```python
plt.imshow(cv.bitwise_or(img1, img2))
```

- ### Bitwise NOT

```python
plt.imshow(cv.bitwise_not(img2))
```

- ### Bitwise XOR

```python
plt.imshow(cv.bitwise_xor(img1, img2))
```

> Hey you ever wonder what bitwise means in this context? Well it's just that, the specific operation (AND, OR, NOT etc.) is performed bitwise (one bit at a time) for each channel in the images, meaning that for color images, it will be applied to each of the Red, Green, and Blue channels separately.
